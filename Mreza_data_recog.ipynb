```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

df = pd.read_csv("group_26_train.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
df.sample(5)
```

```python
df.info()
```

```python
numeric_cols = [col for col in df.select_dtypes(include=[np.number]).columns if col != 'y']
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
bool_cols = df.select_dtypes(include=["bool"]).columns.tolist()

print("Numeric columns:", len(numeric_cols))
print(numeric_cols)

print("\nCategorical columns:", len(categorical_cols))
print(categorical_cols)

print("\nBoolean columns:", len(bool_cols))
print(bool_cols)
```

```python
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100

missing_report = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing Percent": missing_percent
}).sort_values("Missing Percent", ascending=False)

missing_report
```

```python
desc_stats = df[numeric_cols].describe().T

desc_stats["range"] = desc_stats["max"] - desc_stats["min"]

desc_stats
```

```python
skewness = df[numeric_cols].skew().sort_values(ascending=False)

skewness
```

```python
outlier_counts = {}

for col in numeric_cols:
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    
    outlier_counts[col] = outliers

outlier_df = pd.DataFrame.from_dict(outlier_counts, orient="index", columns=["Outlier Count"])

outlier_df.sort_values("Outlier Count", ascending=False)
```

```python
unique_counts = df.nunique().sort_values(ascending=False)

unique_counts
```

```python
constant_cols = [col for col in df.columns if df[col].nunique() <= 1]

print("Constant columns:")
print(constant_cols)
```

```python
near_constant = []

for col in df.columns:
    
    top_freq = df[col].value_counts(normalize=True).values[0]
    
    if top_freq > 0.95:
        near_constant.append(col)

print("Near constant features:")
print(near_constant)
```

```python
target_counts = df["y"].value_counts()
target_ratio = df["y"].value_counts(normalize=True) * 100

target_df = pd.DataFrame({
    "Count": target_counts,
    "Percent": target_ratio
})

target_df
```

```python
duplicate_mask = df.duplicated()

duplicate_rows = df[duplicate_mask]

print("Number of duplicate rows:", duplicate_rows.shape[0])
duplicate_rows
```

```python
zero_counts = (df[numeric_cols] == 0).sum().sort_values(ascending=False)

zero_counts
```

## ۱.۴ شناخت داده (Data Understanding)

### داده شامل چه ستون‌ها یا ویژگی‌هایی است؟
این مجموعه‌داده شامل **۱۰۰,۰۰۰ نمونه و ۳۵ ستون** می‌باشد. ویژگی‌ها را می‌توان به صورت زیر دسته‌بندی کرد:

- **ویژگی‌های زمانی و شناسایی:**  
  شامل ستون منبع داده (`f1`) و زمان ثبت رویداد (`f2`).

- **مختصات جغرافیایی:**  
  ویژگی‌های عددی (`f3` تا `f6`) که عرض و طول جغرافیایی محل وقوع رویداد را نشان می‌دهند.

- **اطلاعات مکانی و آدرس:**  
  ویژگی‌های متنی یا دسته‌ای (`f8` تا `f12`) شامل خیابان، شهر، شهرستان، ایالت و کد پستی.

- **سنجش‌های محیطی و عددی:**  
  متغیرهای عددی (`f7` و `f14` تا `f18` و `f20`) که اندازه‌گیری‌های محیطی یا فیزیکی را نشان می‌دهند، به همراه ویژگی‌های دسته‌ای مرتبط (`f13`، `f19`، `f21`).

- **ویژگی‌های دوحالتی (Boolean):**  
  ستون‌های (`f22` تا `f34`) که مقدار True/False دارند و وجود یا عدم وجود برخی شرایط یا ویژگی‌های مکانی را مشخص می‌کنند.

- **متغیر هدف:**  
  ستون نهایی با نام `y`.

### هر نمونه نشان‌دهنده چیست؟
هر ردیف از داده نشان‌دهنده **ثبت یک رویداد واحد** در یک زمان و مکان مشخص است.  
این رکورد ترکیبی از اطلاعات زمانی، مکانی، محیطی و شرایط مرتبط با آن رویداد را ارائه می‌دهد.

### هدف پروژه بر اساس این داده چیست؟
هدف پروژه، **ساخت یک مدل پیش‌بینی‌کننده** است که بتواند با استفاده از ویژگی‌های موجود، کلاس مربوط به هر رویداد را پیش‌بینی کند (ستون `y`).

### متغیر هدف کدام است؟
متغیر هدف ستون **`y`** است که یک متغیر **دسته‌ای با ۴ کلاس (۱، ۲، ۳ و ۴)** می‌باشد.

### نوع مسئله چیست؟
با توجه به اینکه داده‌ها دارای برچسب هستند و هدف پیش‌بینی یک متغیر دسته‌ای است، این مسئله از نوع **یادگیری نظارت‌شده (Supervised Learning)** و به طور خاص **طبقه‌بندی چندکلاسه (Multi-class Classification)** می‌باشد.
